CLUSTER

```python
#Now at last we will try to find the optimum number of clusters from the data set!
```

```python
ds1=data[['xyz_campaign_id','age','gender','interest','Impressions','Clicks','Spent','Total_Conversion','Approved_Conversion']]
ds1.gender = ds1['gender'].map({'F':1,'M':0})
#replace the range with average values for easier calculations
ds1['age']=ds1['age'].replace(['30-34','35-39','40-44','45-49'],[32,37,42,47])
ds1[['age']] = ds1[['age']].apply(pd.to_numeric) 
ds1.head()
```

```python
#Using Elbow method to find the number of clusters!
from sklearn.cluster import KMeans
%matplotlib inline
wcss=[]
K_rng=10
for i in range(1,K_rng):
    K=KMeans(i)
    K.fit(np.array(ds1))
    w=K.inertia_
    wcss.append(w)
    
Clusters=range(1,K_rng)
plt.figure(figsize=(10,4))
plt.plot(Clusters,wcss)
plt.xlabel('Clusters')
plt.ylabel('WCSS Values') #Within Cluster Sum of Squares
plt.title('Elbow Method Visualisation')
```

```python
#Fitting the model
K3= KMeans(3)
K3.fit(ds1)
#Prediction using the model
ds1_pred2=ds1.copy()
ds1_pred2['Predicted']=K3.fit_predict(ds1)
```

```python
#Visualise the clusters (Impressions and Clicks) after prediction
plt.figure(figsize=(8,5))
plt.scatter(ds1_pred2['Impressions'], ds1_pred2['Clicks'], c=ds1_pred2['Predicted'], cmap = 'jet')
plt.xlabel('Impressions')
plt.ylabel('Clicks')
plt.title('Impressions VS Clicks(K=3)')
```

```python
#Visualise the clusters (Clicks and Conversion) after prediction
plt.figure(figsize=(8,5))
plt.scatter(ds1_pred2['Clicks'], ds1_pred2['Total_Conversion'], c=ds1_pred2['Predicted'], cmap = 'rainbow')
plt.xlabel('Clicks')
plt.ylabel('Conversion')
plt.title('Clicks VS Conversion(K=3)')
```

From the Clustering analysis, 3 clusters are optimal. 
This helps the Company improve. Impressions VS Clicks, Clicks VS Conversion are observed in these cluster segments.
